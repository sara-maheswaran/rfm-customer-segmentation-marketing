## Team Approach
In this project, we began by establishing a foundational understanding of our collective goals and interests. This allowed us to align our vision and set clear expectations for our collaboration.

#### Goals
- **Examine Interests:** Understanding where each team member's interests lie and our industry allowed us to leverage our diverse skills effectively.
- **Build Relationships:** Foster an environment of trust and openness, encouraging everyone to contribute freely.

Our collaborative approach, built on open communication, regular check-ins, and the effective use of digital tools, has fostered a productive working environment. By maintaining a clear focus on our goals and encouraging active participation, we have enhanced our ability to work together successfully on this project.

---

## Team Practices
We emphasized the importance of being open to each other's ideas and perspectives. Aspects of our collaborative approach included:

- **Open Communication:** We were honest about our strengths, challenged each other's work/ideas respectfully. We had several ways to communicate and at times split-off into smaller groups for efficiency.
- **Asking Questions:** We asked questions to clarify doubts and explore ideas further.
- **Presentations:** Sharing our work and insights we found allowed us to incorporate each other's feedback.
- **Standup Meetings:** Regular standup meetings were held to keep everyone on track: we shared updates on our tasks, any roadblocks, and next steps, promoting accountability and teamwork.
